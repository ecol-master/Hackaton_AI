```python
import pandas as pd


with open('points_clean', 'r') as points_clean_file:
    points_clean = points_clean_file.readlines()

secs = []
points = []

for i in range(len(points_clean)-1):
    if points_clean[i].strip().startswith("sec:"):
        temp_sec = points_clean[i].strip()[5:]
        secs.append(temp_sec)
    if points_clean[i].strip().startswith("data:"):
        points.append([points_clean[i+a+1].strip()[2:] for a in range(128)])
print(len(points), len(secs))

df_points = pd.DataFrame(points)
df_points['clean'] = 1
display(df_points.head())
```

```python
with open('points_pit_1', 'r') as points_clean_file:
    points_clean = points_clean_file.readlines()

secs = []
points = []

for i in range(len(points_clean)-1):
    if points_clean[i].strip().startswith("sec:"):
        temp_sec = points_clean[i].strip()[5:]
        secs.append(temp_sec)
    if points_clean[i].strip().startswith("data:"):
        points.append([points_clean[i+a+1].strip()[2:] for a in range(128)])

with open('points_pit_2', 'r') as points_clean_file:
    points_clean = points_clean_file.readlines()

for i in range(len(points_clean)-1):
    if points_clean[i].strip().startswith("sec:"):
        temp_sec = points_clean[i].strip()[5:]
        secs.append(temp_sec)
    if points_clean[i].strip().startswith("data:"):
        points.append([points_clean[i+a+1].strip()[2:] for a in range(128)])


df_points_pit = pd.DataFrame(points)
df_points_pit['clean'] = 0
display(df_points_pit.head())
```

```python
df = pd.concat([df_points, df_points_pit])
print(df.shape)
df.head()
```

```python
df.to_csv('points.csv', index=False)
```
